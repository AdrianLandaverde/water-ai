# 01. Librerías y Datos

```python
import pandas as pd
import plotly.express as px
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly
import geopandas as gpd
```

```python
df_sequia= pd.read_excel('../data/Municipios_con_ sequia.xlsx', sheet_name=2)
df_sequia
```

```python
gdf_estados= gpd.read_file('../data/dest2019gw/dest2019gw.shp')
gdf_estados
```

# 02. Limpieza de datos

```python
df_sequia.columns
columnas_fechas= list(df_sequia.columns[9:])
df_sequia= df_sequia[ ['CVE_CONCATENADA','CVE_ENT','CVE_MUN','NOMBRE_MUN','ENTIDAD'] + columnas_fechas]
df_sequia
```

```python
df_sequia= df_sequia.melt(id_vars=['CVE_CONCATENADA','CVE_ENT','CVE_MUN','NOMBRE_MUN','ENTIDAD'], var_name='Fecha', value_name='Sequia')
df_sequia['Sequia']= df_sequia['Sequia'].replace({'D0': 0, 'D1': 1, 'D2': 2, 'D3': 3, 'D4': 4, 'D5': 5})
df_sequia['Sequia']= df_sequia['Sequia'].fillna(0)
df_sequia['Sequia']= df_sequia['Sequia'].astype(int)
df_sequia
```

```python
df_sequia['Fecha']= pd.to_datetime(df_sequia['Fecha'], format='%Y-%m-%d')
df_sequia['mes']= df_sequia['Fecha'].dt.month
df_sequia['anio']= df_sequia['Fecha'].dt.year
df_sequia_estado= df_sequia.groupby(['CVE_ENT','ENTIDAD','anio', 'mes'])[['Sequia']].mean().reset_index()
df_sequia_estado['Fecha']= pd.to_datetime(df_sequia_estado['anio'].astype(str) + '-' + df_sequia_estado['mes'].astype(str) + '-1')
df_sequia_estado
```

```python
gdf_estados['CVE_ENT']= gdf_estados['CVE_ENT'].astype(int)
gdf_estados_sequias= gdf_estados.merge(df_sequia_estado, on='CVE_ENT', how='left')
gdf_estados_sequias
```

```python
gdf_estados_sequias.dtypes
```

# 03. Visualizar Datos

```python
gdf_estados_sequias['anio']
```

```python
gdf_plot= gdf_estados_sequias[gdf_estados_sequias['anio']>=2023].copy()
gdf_plot= gdf_plot.reset_index(drop=True)
import json
from shapely.geometry import mapping


# Convert the 'geometry' column from geometry objects to GeoJSON
gdf_plot['geometry'] = gdf_plot['geometry'].apply(lambda x: json.dumps(mapping(x)))

# Create a GeoJSON object
geojson = {
    'type': 'FeatureCollection',
    'features': [
        {
            'type': 'Feature',
            'geometry': json.loads(geom),
            'id': str(index),
        }
        for index, geom in gdf_plot['geometry'].items()
    ]
}

fig = px.choropleth(gdf_plot, 
                    geojson=geojson, 
                    locations=gdf_plot.index.astype(str), 
                    color='Sequia',  
                    hover_name='ENTIDAD', 
                    title='Sequias en México',
                    color_continuous_scale=px.colors.diverging.Temps,
                    animation_frame='Fecha')

fig.update_geos(fitbounds="locations", visible=False)

fig.write_html('mapa_sequias.html')
```

# 04. Modelos de Prophet

```python
df_estado= df_sequia_estado[df_sequia_estado['CVE_ENT']==10].copy()
df_estado= df_estado.rename(columns={'Fecha': 'ds', 'Sequia': 'y'})
df_estado= df_estado[['ds', 'y']]
m= Prophet(changepoint_prior_scale=1,seasonality_mode='multiplicative', seasonality_prior_scale=10)
m.fit(df_estado)
future= m.make_future_dataframe(periods=12, freq='ME')
forecast= m.predict(future)
```

```python
fig= plot_plotly(m, forecast, changepoints=True)
fig.update_layout(title='Sequias en Yucatán', xaxis_title='Fecha', yaxis_title='Sequia')
```

```python
plot_components_plotly(m, forecast)
```
